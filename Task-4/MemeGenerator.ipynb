```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
img = Image.open(r"/content/meme_5.jpg")
```

```python
plt.imshow(img)
plt.axis('on')
plt.show()
```

```python
pen = ImageDraw.Draw(img)
```

```python
elite_font = ImageFont.truetype(r"/content/CrimsonText-Regular.ttf", 90)
```

```python
pen.text((700, 100), "\nFix the code", font=elite_font, fill=(0, 0, 0))  # Black text for top
pen.text((650, 600), "\nBlame the\nprogramming \nlanguage", font=elite_font, fill=(0, 0, 0))  # Black text for bottom


plt.imshow(img)
plt.axis('on')
plt.show()
plt.savefig('saved_image.png')
```
